# 1. Calculate Image gradient of tiger image (20 points)

- read and show the tiger image
- show the gradient of x and y, respectively

```python
#Read image from GIT location and store it into a variable
image_url = "https://github.com/ankitagg2008/DLNN/raw/2c3e914eef92980526e1f97303aa76bc0d8ec4fb/images/tiger.png"
response = requests.get(image_url)
org_image = cv2.imdecode(np.frombuffer(response.content, np.uint8), cv2.IMREAD_COLOR)
org_gray = cv2.cvtColor(org_image[:,:,::-1], cv2.COLOR_BGR2GRAY)

#Defind subplots size
fig, axes = plt.subplots(3, 2, figsize=(12, 12))

# Resize / Change its dimensions and display
tiger_resized_image = cv2.resize(org_gray,(480, 320)) ;

#Define Mask and convert datatype to make it uniform. 
mask = np.array([[-1, 1]], dtype=np.float32)

# Compute the image gradient in the x direction
derX = cv2.filter2D(tiger_resized_image.astype(np.float32), -1, mask)

# Compute the image gradient in the y direction
derY = cv2.filter2D(tiger_resized_image.astype(np.float32), -1, mask.T)

# Compute the image gradient magnitude
gmag = np.sqrt(derX**2 + derY**2)

# Display the Original, Resized, dx and dy images
axes[0,0].imshow(org_image);axes[0,0].set_title("Original Image")
axes[0,1].imshow(tiger_resized_image, cmap='gray');axes[0,1].set_title("Tiger Resized Image (480, 320)")
axes[1,0].imshow(derX, cmap='gray');axes[1,0].set_title("X-Gradient")
axes[1,1].imshow(derY, cmap='gray');axes[1,1].set_title("Y-Gradient")
axes[2,0].imshow(gmag, cmap='gray');axes[2,0].set_title("Gradient Magnitude")

fig.delaxes(axes[2, 1])
```

# 2. Implement Sobel edge detection of bird image (30 points)

- read and show the bird image
- show the gradient of x and y, respectively
- show gradient magnitude
- show results of different thresholds


```python
image_url = "https://github.com/ankitagg2008/DLNN/raw/2c3e914eef92980526e1f97303aa76bc0d8ec4fb/images/bird.png"
response = requests.get(image_url)
org_image = cv2.imdecode(np.frombuffer(response.content, np.uint8), cv2.IMREAD_COLOR)
org_image = cv2.cvtColor(org_image, cv2.COLOR_BGR2GRAY)
```

```python
#Defind subplots size
fig, axes = plt.subplots(3, 2, figsize=(15, 15))

# Resize / Change its dimensions and display
bird_resized_image = cv2.resize(org_image,(480, 320)) 

# Compute x and y gradients using Sobel function of CV2
gx = cv2.Sobel(org_image, cv2.CV_64F, 1, 0, ksize=5)
gy = cv2.Sobel(org_image, cv2.CV_64F, 0, 1, ksize=5)

# Normalize the gradients to the range [0, 1]
gx_normalized = (gx + 4) / 8
gy_normalized = (gy + 4) / 8

# code to find the gradient magnitude and direction using CV2 function
gmag, gdir = cv2.cartToPolar(gx, gy, angleInDegrees=True)
gmag_normalized = gmag / (4 * np.sqrt(2))

# Normalize the gradient direction to [0, 1]
gdir_normalized = (gdir + 180.0) / 360.0

# Display the Original, Resized, dx and dy images
axes[0,0].imshow(org_image, cmap='gray');axes[0,0].set_title("Original Image")
axes[0,1].imshow(bird_resized_image, cmap='gray');axes[0,1].set_title("Bird Resized Image (480, 320)")
axes[1,0].imshow(gx_normalized, cmap='gray');axes[1,0].set_title("X-Gradient")
axes[1,1].imshow(gy_normalized, cmap='gray');axes[1,1].set_title("Y-Gradient")
axes[2,0].imshow(gmag_normalized, cmap='gray');axes[2,0].set_title("Gradient Magnitude")
axes[2,1].imshow(gdir_normalized, cmap='gray');axes[2,1].set_title("Gradient Direction")
```

```python
# defining subplot size
fig, axes = plt.subplots(3, 3, figsize=(12, 9))
count = 0 

# for loop to display images with different thresholds
for i in range(50, 1700, 200):
    k = count // 3  # Calculate the row index
    l = count % 3   # Calculate the column index
    
    binary_image1 = gmag > i
    axes[k, l].imshow(binary_image1, cmap='gray')
    axes[k, l].set_title('Threshold value: ' + str(i))
    
    count += 1
```

# 3. Implement LOG edge detection of bird image (20 points)

- read and show the bird image
- show results of LOG edge detection

```python
image_url = "https://github.com/ankitagg2008/DLNN/raw/2c3e914eef92980526e1f97303aa76bc0d8ec4fb/images/bird.png"
response = requests.get(image_url)
org_image = cv2.imdecode(np.frombuffer(response.content, np.uint8), cv2.IMREAD_COLOR)
org_image = cv2.cvtColor(org_image, cv2.COLOR_BGR2GRAY)

#Defind subplots size
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Resize / Change its dimensions and display
bird_resized_image = cv2.resize(org_image,(480, 320)) 

h = cv2.getGaussianKernel(7, 1)
h = np.outer(h, h)

# Apply the LOG filter
imFiltered = cv2.Laplacian(bird_resized_image.astype(np.float64),cv2.CV_64F, ksize = 5)

# Create a binary image using a threshold
threshold = 2
binaryImage = imFiltered > threshold

# Display the Original, Resized, dx and dy images
axes[0,0].imshow(org_image, cmap='gray');axes[0,0].set_title("Original Image")
axes[0,1].imshow(bird_resized_image, cmap='gray');axes[0,1].set_title("Bird Resized Image (480, 320)")
axes[1,0].imshow(imFiltered, cmap='gray');axes[1,0].set_title("Log Filtered Image")
axes[1,1].imshow(binaryImage, cmap='gray');axes[1,1].set_title("Binary Image (Threshold=2)")

```

##### 5. Given Gaussian function: $g(x,y) = \frac{1}{\sqrt{2\pi}\sigma} e^{(-\frac{x^2 + y^2}{2\sigma^2})}$ , calculate $\frac{\partial^2 g}{\partial x^2} + \frac{\partial^2 g}{\partial y^2}$ 
# (10 points)

(You can search for how to use LaTex to write equations if you don't know how to write.)

### Solution Approach

   1. Given function $g(x,y)$, we need to calculate the sum of the second order particial derivative w.r.t. $x$ and $y$.
   2. First calculate the first derivative $\frac{\partial g}{\partial x}$ for function $g(x,y)$ then perform one more derivation to obtain the second order derivative $\frac{\partial^2 g}{\partial x^2}$.
   3. First calculate the first derivative $\frac{\partial g}{\partial y}$ for function $g(x,y)$ then perform one more derivation to obtain the second order derivative $\frac{\partial^2 g}{\partial y^2}$.
   4. Then add them both $\frac{\partial^2 g}{\partial x^2} + \frac{\partial^2 g}{\partial y^2}$.

#### Given function
$$g(x,y) = \frac{1}{\sqrt{2\pi}\sigma} e^{(-\frac{x^2 + y^2}{2\sigma^2})}$$

#### Calculate 

$$\frac{\partial^2 g}{\partial x^2} + \frac{\partial^2 g}{\partial y^2}$$ 

#### Find first order derivative w.r.t. $x$

$$= \frac{\partial}{\partial x} \left[\frac{1}{\sqrt{2\pi}\sigma} e^{(-\frac{x^2 + y^2}{2\sigma^2})}\right]$$

$$= \frac{1}{\sqrt{2\pi}\sigma} \frac{d}{dx}\left[ e^{\frac{y^2 - x^2}{2\sigma^2}} \right]$$

$$= e^{\frac{y^2 - x^2}{2\sigma^2}} \cdot \frac{d}{dx}\left[ \frac{y^2 - x^2}{2\sigma^2} \right] \frac{1}{\sqrt{2\pi}\sigma}$$

$$= -e^{\frac{y^2 - x^2}{2\sigma^2}} \cdot \frac{x}{\sqrt{2\pi}\sigma^3}$$

#### Now using above equation perform one more derivative to obtain second level derivative w.r.t. $x$

$$\frac{d}{dx}\left[-e^{\frac{y^2 - x^2}{2\sigma^2}} \cdot \frac{x}{\sqrt{2\pi}\sigma^3}\right]$$

$$= -\frac{1}{\sqrt{2\pi}\sigma^3} \cdot \frac{d}{dx}\left[ x e^{\frac{y^2 - x^2}{2\sigma^2}} \right]$$

$$= -\frac{d}{dx}\left[ x \right] \cdot e^{\frac{y^2 - x^2}{2\sigma^2}} - x \cdot \frac{d}{dx}\left[ e^{\frac{y^2 - x^2}{2\sigma^2}} \right] \cdot \frac{1}{\sqrt{2\pi}\sigma^3}$$

$$= \left(-e^{\frac{y^2 - x^2}{2\sigma^2}} - \frac{x^2e^{\frac{y^2 - x^2}{2\sigma^2}}}{\sigma^2}\right) \cdot \frac{1}{\sqrt{2\pi}\sigma^3}$$

#### This can be arrange and rewrite as: 

$$\frac{\partial^2 g}{\partial x^2} = \left(x^2 - \sigma^2\right)e^{\frac{y^2 - x^2}{2\sigma^2}} \cdot \frac{1}{\sqrt{2\pi}\sigma^5}$$



#### Similarly, second level derivative w.r.t. $y$ is equal to

$$\frac{\partial^2 g}{\partial y^2} = \left(y^2 + \sigma^2\right)e^{\frac{y^2 - x^2}{2\sigma^2}} \cdot \frac{1}{\sqrt{2\pi}\sigma^5}$$


#### Add both the second order partial derivative

$$\frac{\partial^2 g}{\partial x^2} + \frac{\partial^2 g}{\partial y^2} = \left(x^2 - \sigma^2\right)e^{\frac{y^2 - x^2}{2\sigma^2}} \cdot \frac{1}{\sqrt{2\pi}\sigma^5} +  \left(y^2 + \sigma^2\right)e^{\frac{y^2 - x^2}{2\sigma^2}} \cdot \frac{1}{\sqrt{2\pi}\sigma^5} $$


$$\frac{\partial^2 g}{\partial x^2} + \frac{\partial^2 g}{\partial y^2} = \frac{1}{\sqrt{2\pi}\sigma^5} \left(x^2 +y^2 -2\sigma^2\right) e^{(-\frac{x^2 + y^2}{2\sigma^2})}$$

### Solution 


$$\frac{\partial^2 g}{\partial x^2} + \frac{\partial^2 g}{\partial y^2} = \frac{1}{\sqrt{2\pi}\sigma^5} \left(x^2 +y^2 -2\sigma^2\right) e^{(-\frac{x^2 + y^2}{2\sigma^2})}$$
