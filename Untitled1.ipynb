## Dataset

In order to develop the recommender system, we will utilize the publicly available books-crossing dataset on Kaggle.com

Dataset Link: https://www.kaggle.com/datasets/arashnic/book-recommendation-dataset/data

Dataset Details: 

The Book-Crossing dataset comprises 3 csv files.

   - **Users**
       
       Users CSV file consists of data related to the users who has provided raiting to the books. It consists of following fields: 
       - User-ID
       - Location
       - Age


   - **Books**
        
        Books CSV file consists of details about a particular book. The information available in this file is obtained from Amazon Web Services. 
        - Book-Title
        - Book-Author
        - Year-Of-Publication
        - Publisher
        - Image-URL-,
        - Image-URL-M
        - Image-URL-L


   - **Ratings**
   
        Rating CSV file conisists of the book rating information. 
        
        - Book-Rating expressed on a scale from 1-10 (Explicit) or Implicit, expressed by 0.
